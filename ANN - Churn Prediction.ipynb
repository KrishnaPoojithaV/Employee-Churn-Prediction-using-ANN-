```python
le = LabelEncoder()
df['Geography'] = le.fit_transform(df['Geography'])
df['Gender'] = le.fit_transform(df['Gender'])
df.head()
```

```python
x = df.iloc[:,3:-1].values
y= df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z = one.fit_transform(x[:,1:2]).toarray()
x = np.delete(x,1,axis = 1)
x=np.concatenate((z,x),axis=1)
```

```python
x
```

```python
x = x[:,1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units=11, init='uniform', activation='relu'))
```

```python
classifier.add(Dense(units=6, init='uniform', activation='relu'))
```

```python
classifier.add(Dense(units=1, init='uniform', activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(x_train,y_train, batch_size=32, epochs=2)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(" ")
print(classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```
